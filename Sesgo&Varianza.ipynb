**Principales apuntes e Interpretaciones**

*   **Fuentes de Sesgo** la fuente de informacion en la medida que no sea representativa en terminos estadisticos puede ser una gran fuente de sesgo, por lo que idealmente debemos buscar varias fuentes o un fuenta que posea bastante representacion sobre la variable de estudio.

*   **Fuentes de Varianza** Por otra parte una muestra de bajo volumen puede ser una gran fuente de error por varianza, debido a que los resultados pueden ser inconsistentes con el universo total, debido al bajo nivel de representacion, por lo que debemos buscar un Nº optimo de muestra

* **Equilibrio entre sesgo y varianza**: en base a los 2 elementos antes expuestos, deberemos buscar el adecuado intercambio entre sesgo y varianza que permitan mantener un adecuado nivel de error para ambos.



**Gestionando el Sesgo y Varianza**

*Mantener tus instintos:* Gran parte de las personas suelen sostener que es preferible aumentar el nivel de error por varianza que por sesgo, debido a que un error por sesgo implica que tanto su modelo como algoritmo esta basicamente erroneo, esto suele interpretarse en modelos con bajos nivel de muestras, sin embargo en la practica lo importante es el desempeño del modelo que tu posees, por lo que la gestion de ambos errores es necesaria por igual.

*Embolsado y Remuestreo* estas tecnicas pueden ser utilizadas para reducir el nivel de varianza en modelos de prediccion. en embolsado se puede replicar mucha informacion desde el conjunto de datos original usando una seleccion aleatoria con reemplazo. un poderoso modelo de algoritmo es "Bosque aleatorio" este trabaja entrenando numerosos arboles de decision cada uno basado en un diferente embolsado de la informacion original. al crear muchos de estos arboles, como efecto bosque y promediando la varianza del modelo final puede reducirlo significativamente a un solo arbol.

*Propiedades asintóticas de los algoritmos*: en terminos sencillos esto implica que si el tamaño de la muestra de entrenamiento crece al infinito, el error por sesgo de tu modelo caera a (consistencia asintótica) y tu modelo tendra una varianza que no sera peor que otro potencial modelo que tu hayas podido utilizar (eficiencia asintótica)

*Comprender el ajuste excesivo y el ajuste insuficiente*: en esencia, lidiar con el sesgo y la varianza se trata de lidiar con el ajuste excesivo y el insuficiente. el sesgo se reduce y la varianza aumenta en relacion con la complejidad del modelo. a medida que se agregan mas y mas parametro a un modelo , la complejidad del modelo aumenta y la varianza se convierte en nuestra principal preocupacion, mientras el sesgo disminuye constantemente.
